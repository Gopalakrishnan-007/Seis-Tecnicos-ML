```python
# Create a list to store feature data
data = []

# Iterate through each file
for dirname, _, filenames in os.walk('/kaggle/input/gun-shot/38sws_dot38_caliber'):
    for filename in filenames:
        file_path = os.path.join(dirname, filename)
        
        # Load the audio file
        audio_data, sample_rate = librosa.load(file_path, sr=None)
        
        # Extract MFCC features
        mfccs = librosa.feature.mfcc(y=audio_data, sr=sample_rate, n_mfcc=13)
        mfccs_mean = mfccs.mean(axis=1)
        
        # Append the data
        data.append([filename] + mfccs_mean.tolist())

# Convert the data into a DataFrame
df = pd.DataFrame(data, columns=['filename'] + [f'mfcc_{i}' for i in range(1, 14)])

# Save to CSV
df.to_csv('audio_features.csv', index=False)
```

```python
# Set to store sample rates and durations
sample_rates = set()
durations = set()

# Iterate through each file
for dirname, _, filenames in os.walk('/kaggle/input/gun-shot/38sws_dot38_caliber'):
    for filename in filenames:
        file_path = os.path.join(dirname, filename)
        
        # Load the audio file
        audio_data, sample_rate = librosa.load(file_path, sr=None)
        duration = librosa.get_duration(y=audio_data, sr=sample_rate)
        
        # Add sample rate and duration to the sets
        sample_rates.add(sample_rate)
        durations.add(duration)

# Display the sample rates and durations found
print(f"Sample Rates: {sample_rates}")
print(f"Durations: {durations}")
```

```python
# Define the target duration (in seconds, based on the longest file in your dataset)
target_duration = max(durations)  # or set a specific duration like 8.0 seconds

# Function to pad the audio to the target duration
def pad_audio(audio_data, sample_rate, target_duration):
    target_length = int(target_duration * sample_rate)
    if len(audio_data) < target_length:
        # Calculate the difference and pad with zeros (silence)
        padding = target_length - len(audio_data)
        padded_audio = np.pad(audio_data, (0, padding), 'constant')
    else:
        # If longer, trim the audio
        padded_audio = audio_data[:target_length]
    return padded_audio

# Iterate through each file
for dirname, _, filenames in os.walk('/kaggle/input/gun-shot/38sws_dot38_caliber'):
    for filename in filenames:
        file_path = os.path.join(dirname, filename)
        
        audio_data, sample_rate = librosa.load(file_path, sr=None)
        padded_audio = pad_audio(audio_data, sample_rate, target_duration)
```

```python
import soundfile as sf  # To save the audio files
```

```python
# Define the target duration (in seconds, based on the longest file in your dataset)
target_duration = max(durations)  # or set a specific duration like 8.0 seconds

# Create an output directory if it doesn't exist
output_dir = '/kaggle/working/38sws_dot38_caliber/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Function to pad the audio to the target duration
def pad_audio(audio_data, sample_rate, target_duration):
    target_length = int(target_duration * sample_rate)
    if len(audio_data) < target_length:
        # Calculate the difference and pad with zeros (silence)
        padding = target_length - len(audio_data)
        padded_audio = np.pad(audio_data, (0, padding), 'constant')
    else:
        # If the audio is already the correct length or longer, return it as is
        padded_audio = audio_data
    return padded_audio

# Iterate through each file
for dirname, _, filenames in os.walk('/kaggle/input/gun-shot/38sws_dot38_caliber'):
    for filename in filenames:
        file_path = os.path.join(dirname, filename)
        
        # Load the audio file
        audio_data, sample_rate = librosa.load(file_path, sr=None)
        
        # Pad the audio
        padded_audio = pad_audio(audio_data, sample_rate, target_duration)
        
        # Construct the output file path
        output_file_path = os.path.join(output_dir, f"{os.path.splitext(filename)[0]}_padded.wav")
        
        # Save the padded audio file
        sf.write(output_file_path, padded_audio, sample_rate)

        print(f"Saved padded audio: {output_file_path}")
```

```python
df1 = pd.read_csv("/kaggle/input/gunshot-label-dataset/gunshot-audio-all-metadata.csv")

# Define the file to check
file_to_check = "03fc4685-909e-42c5-aff0-f519f1d14b12_chan0_v0"

# Check if the file exists in the DataFrame
# Assuming there's a column in the CSV that contains the filenames (e.g., "filename" or similar)
if (df1['filename'] == file_to_check).any():
    print(f"{file_to_check} exists in the metadata.")
else:
    print(f"{file_to_check} does not exist in the metadata.")
```

```python
# List to store the features for each file
data = []

# Iterate through each audio file
for dirname, _, filenames in os.walk('/kaggle/working/38sws_dot38_caliber'):
    for filename in filenames:
        file_path = os.path.join(dirname, filename)
        
        # Load the audio file
        audio_data, sample_rate = librosa.load(file_path, sr=None)
        
        # Pad the audio data to ensure consistent length
        padded_audio = pad_audio(audio_data, sample_rate, target_duration)
        
        # Extract Mel Spectrogram
        mel_spectrogram = librosa.feature.melspectrogram(y=padded_audio, sr=sample_rate, n_mels=128)
        mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
        mel_spectrogram_mean = mel_spectrogram_db.mean(axis=1)
        
        # Extract MFCCs
        mfccs = librosa.feature.mfcc(y=padded_audio, sr=sample_rate, n_mfcc=13)
        mfccs_mean = mfccs.mean(axis=1)
        
        # Combine features and metadata
        features = [filename] + mfccs_mean.tolist() + mel_spectrogram_mean.tolist()
        data.append(features)

# Create a DataFrame for the extracted features
columns = ['filename'] + [f'mfcc_{i}' for i in range(1, 14)] + [f'mel_{i}' for i in range(1, 129)]
df = pd.DataFrame(data, columns=columns)

# Save to CSV for future use
df.to_csv('audio_features.csv', index=False)
```
